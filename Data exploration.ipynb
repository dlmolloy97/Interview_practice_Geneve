```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
#Here, I import the data and obtain the needed columns from the dataframe
listings = pd.read_csv('data/listings.csv')
listings.head()
df = listings[['latitude','longitude','price','minimum_nights']]
```

```python
#Reference: https://pythonbasics.org/what-is-flask-python/
map = folium.Map(location=[df.latitude.mean(), df.longitude.mean()], zoom_start=14, control_scale=True)
for index, location_info in df.iterrows():
    folium.Circle([location_info["latitude"], location_info["longitude"]], 
                  color = location_info['price'],
                        popup=location_info["price"], radius = 0, weight = 3, fill = True, fill_opacity=1).add_to(map)
map
```
